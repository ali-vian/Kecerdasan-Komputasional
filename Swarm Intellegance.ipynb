## PSO

- partikel = burung
- bergerak berdasarkan inertia,cognitive,social
- yang terbaik ditentukan dengan fungsi fitness
- PSO partikel di pengaruhi oleh melihat diri sendiri, histori, dan teman-teman kalau GA hanya dipengaruhi diri sendiri

- c = diset sama semua jika ingin tidak ada yang dominan, jika ingin lebih dominan inertia maka c0 diset lebih tinggi
- r1,r2 = element stochastic dibangkitkan secara random
- pi/p_best = posisi terbaik dari partikel
- pg/g_best = posisi terbaik dari semua partikel

#### P best

jika f(xi(t)) >= f(pi(t-1)) maka lebih kecil f(pi(t-1)) maka p_best sebelumnya
jika f(xi(t)) < f(pi(t-1)) maka lebih kecil f(pi(t-1)) maka p_best sekarang

- pg = min(p0(t),p1(t)...pn(t))



```python
x = [-6,4,-2,-1,2]
v = [0.1,0.1,0.5,0.2,0.7]
c = [1,1,1]
r = [0.1,0.2]
pi = x.copy()

print("Initialisasi : ")
print(x)
print(v)
def fitness(i):
    return i**2+5*i+10

for j in range(2) :
    temp=[]
    print("\nFitness :")
    for k in range(len(x)):
        print(f"f(p{k+1}({j})) = ",fitness(x[k]))
        if fitness(x[k]) <= fitness(pi[k]):
            pi[k]=x[k]
        temp.append(fitness(x[k]))
    pg = pi[temp.index(min(temp))]
    print("Gbest = ",min(temp),"=>",pg)
    print("\nUpdate Velocity :")
    for i in range(len(x)):
        v[i] = c[0]*v[i]+c[1]*r[0]*(pi[i]-x[i])+c[2]*r[1]*(pg-x[i])
        print(f"v{i+1}({j+1}) = ",v[i])
    print("\nUpdate Partikel :")
    for h in range(len(x)):
        x[h]=x[h]+v[h]
        print(f"x{h+1}({j+1}) = ",x[h])
print("Gbest = ",pg)
```

```python
partikel = [[0,1,2,3,4],
     [4,1,2,0,3],
     [4,3,2,0,1],
     [1,4,2,3,0],
     [3,2,1,4,0],
     [2,1,3,4,0]]
koor  = [[12,6],[2,8],[10,14],[4,2],[6,2]]

import math

for i,x in enumerate(partikel) :
    print(f"x{i+1} = n",x)

def jarak_euclidean(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

def hitung_fitness(rute, koordinat):
    total_jarak = 0
    for i in range(len(rute) - 1):
        total_jarak += jarak_euclidean(koordinat[rute[i]], koordinat[rute[i + 1]])
    total_jarak += jarak_euclidean(koordinat[rute[-1]], koordinat[rute[0]])  # kembali ke kota awal
    return total_jarak

hasil = []
for i in partikel :
    hasil.append(hitung_fitness(i, koor))
print()
for i, fitness in enumerate(hasil):
    print(f"F(x{i+1}): {fitness:.2f}")
    
```
