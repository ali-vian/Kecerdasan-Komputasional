```python
import math

class MembershipDegree:
    def __init__(self):
        self.mu = 0
        
    def panas(self,x,alfa,beta):
        if x<= alfa:
            self.mu = 0
        elif alfa< x <= beta:
            self.mu = (x-alfa)/(beta-alfa)
        else:
            self.mu = 1
            
    def S(self,x,alfa,beta,gama):
        if x <= alfa:
            self.mu = 0
        elif alfa <= x <= beta :
            self.mu = 2*((x-alfa)/(gama-alfa))**2
        elif beta < x <= gama :
            self.mu = 1 - 2*((x-gama)/(gama-alfa))**2
        elif x > gama:
            self.mu = 1
            
    def dingin(self,x,alfa,beta):
        if x <= alfa:
            self.mu = 1 
        elif alfa < x <= beta:
            self.mu = (beta - x) / (beta - alfa) 
        else:
            self.mu = 0  
            
    def triangularFn(self,x,alfa,beta,gama):
        if x <= alfa:
            self.mu = 0
        elif alfa < x and x <= beta:
            self.mu = (x-alfa)/(beta-alfa)
        elif beta<x and x<=gama:
            self.mu = (gama-x)/(gama-beta)
        elif x > gama:
            self.mu = 0
            
    def trapesium(self,x,alfa,beta,gama,delta):
        if x <= alfa:
            self.mu = 0
        elif alfa < x <= beta:
            self.mu = (x-alfa)/(beta-alfa)
        elif beta < x <= gama:
            self.mu = 1
        elif gama < x <= delta:
            self.mu = (delta-x)/(delta-gama)
        else:
            self.mu = 0
            
    def G(self,x,m,sigma):
        self.mu = math.e**((-(x-m)**2)/(2*sigma**2))
            
    def triangularFnCom(self,x,alfa,beta,gama):
        if x <= alfa:
            self.mu = 1
        elif alfa < x and x <= beta:
            self.mu = 1 - ( (x-alfa)/(beta-alfa))
        elif beta<x and x<=gama:
            self.mu = 1 - ((gama-x)/(gama-beta))
        else:
            self.mu = 1
            
    def irisan(self,x,alfa,beta,gama,alfa1,beta1,gama1):
        fuzzy = MembershipDegree()
        fuzzy.triangularFn(x,alfa,beta,gama)
        a = fuzzy.mu
        fuzzy.triangularFnCom(x,alfa1,beta1,gama1)
        b = fuzzy.mu
        self.mu = min(a,b)
```

```python
import numpy as np 
import matplotlib.pyplot as plt
fuzzy = MembershipDegree()
x = np.linspace(0,60,200)
y = np.zeros_like(x)

for i in range(x.shape[0]):
      fuzzy.triangularFn(x[i],20,30,40)
      y[i] = fuzzy.mu
plt.subplot(3, 2, 1)
plt.plot(x,y,'-c')

for i in range(x.shape[0]):
    fuzzy.dingin(x[i],10,20)
    y[i] = fuzzy.mu
plt.subplot(3, 2, 2)
plt.plot(x,y,'-b')

for i in range(x.shape[0]):
    fuzzy.panas(x[i],30,40)
    y[i] = fuzzy.mu
plt.subplot(3, 2, 3)
plt.plot(x,y,'-r')

for i in range(x.shape[0]):
    fuzzy.S(x[i],10,25,40)
    y[i] = fuzzy.mu
plt.subplot(3, 2, 4)
plt.plot(x,y,'-m')

for i in range(x.shape[0]):
    fuzzy.G(x[i],30,5)
    y[i] = fuzzy.mu
plt.subplot(3, 2, 5)
plt.plot(x,y,'-g')

for i in range(x.shape[0]):
    fuzzy.trapesium(x[i],10,20,40,50)
    y[i] = fuzzy.mu
plt.subplot(3, 2, 6)
plt.plot(x,y,'-y')

plt.show()
```
