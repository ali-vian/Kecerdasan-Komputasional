```python
import numpy as np
import skfuzzy
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#  Antecedent
nilai = np.linspace(0,4,200)
pendapatan = np.linspace(0,20,200)

# Consequent
beasiswa = np.linspace(0,3,300)
```

```python
# membership
nilaiRendah = skfuzzy.trimf(nilai,[0,0,2])
nilaiSedang = skfuzzy.trimf(nilai,[1.5,2.5,3.5])
nilaiTinggi = skfuzzy.trimf(nilai,[3,4,4])
plt.plot(nilai,nilaiRendah,"-b")
plt.plot(nilai,nilaiSedang,"-r")
plt.plot(nilai,nilaiTinggi,"-g")
plt.show()

pendapatanRendah = skfuzzy.trapmf(pendapatan,[0,0,2,4])
pendapatanCukup = skfuzzy.trapmf(pendapatan,[2,6,9,13])
pendapatanTinggi = skfuzzy.trapmf(pendapatan,[10,12,20,20])
plt.plot(pendapatan,pendapatanRendah,"-b")
plt.plot(pendapatan,pendapatanCukup,"-r")
plt.plot(pendapatan,pendapatanTinggi,"-g")
plt.show()

beasiswaSedikit = skfuzzy.trapmf(beasiswa,[0,0,1,2])
beasiswaBanyak = skfuzzy.trapmf(beasiswa,[1,2,3,3])
plt.plot(beasiswa,beasiswaSedikit,"-b")
plt.plot(beasiswa,beasiswaBanyak,"-r")
plt.show()
```

```python
# Misalkan data crips
nilaiMhs = np.array([3.5])
pendapatanOrtu = np.array([2.5])

# Fuzzifikasi
mu_nilaiRendah = skfuzzy.trimf(nilaiMhs,[0,0,2])
mu_nilaiSedang = skfuzzy.trimf(nilaiMhs,[1.5,2.5,3.5])
mu_nilaiTinggi = skfuzzy.trimf(nilaiMhs,[3,4,4])

mu_pendapatanRendah = skfuzzy.trapmf(pendapatanOrtu,[0,0,2,4])
mu_pendapatanCukup = skfuzzy.trapmf(pendapatanOrtu,[2,6,9,13])
mu_pendapatanTinggi = skfuzzy.trapmf(pendapatanOrtu,[10,12,20,20])

# Rules
R1 = min( mu_nilaiRendah , mu_pendapatanRendah ) #sedikit
R2 = min( mu_nilaiSedang , mu_pendapatanCukup ) #sedikit
R3 = min( mu_nilaiTinggi , mu_pendapatanCukup ) #sedikit
R4 = min( mu_nilaiTinggi , mu_pendapatanRendah ) #banyak

print(R1)
print(R2)
print(R3)
print(R4)

x = max(R1[0],R2[0],R3[0])

himpBeasiswa = np.arange(0,3.01,0.5)
himpSedikit = np.array([ x if i < 2 else 0 for i in himpBeasiswa])
himpBanyak = np.array([R4[0] if i > 1 else 0 for i in himpBeasiswa])

# Agregasi
agregat = np.fmax(himpSedikit,himpBanyak)

# Defuzzifikasi
y = sum(himpBeasiswa * agregat )/sum(agregat)
print("Hasil : ",y)
```
