```python
x =  np.linspace(0,40,300)
sedang =skfuzzy.trimf(x,[15,25,35])
plt.plot(x,sedang,'-y')

rendah=skfuzzy.trimf(x,[0,0,20])
plt.plot(x,rendah,'-b')

rendah=skfuzzy.trimf(x,[30,40,40])
plt.plot(x,rendah,'-g')
```
